## Test Data
I generated data of an analog phone dialing numbers. 
This is because it's simple data to play with: pressing a digit on an analog phone generated two sine waves simultaneously.

For example, when '1' is pressed, a sine wave at 697hz is created along with one at 1209hz. We will observe the superposition of these two waves in our data.

697hz means that the sine wave completes it's full cycle 697 times within one second. This means if we want to see one cycle, we have to zoom into a 1 second interval 697x!

Our generated audio will be 7 seconds in total.
The first 3 being the sound made when '1' is pressed, the next 1 second being silence, and the last 3 seconds being '2' pressed.

That means first there will be 3 seconds of sine waves at 697hz and 1209hz. Then 1 seconds of nothing. Then 3 seconds of sine waves at 697hz and 1336hz.

We will assume a sampling rate of 4000. This means that every second, we take 4000 samples of the audio. In our total time of 7 seconds, there are then be 4000 * 7 = 28000 samples.

## Data Generation Code
Importing required modules

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning) 
```

First, we define get_signal_Hz. This function outputs a sine wave of a requested length and frequency. 

To do so, it takes a frequency (Hz), the sampling rate (sample_rate), and length (length_ts_sec) as input. It returns a list representing the requested sample. 

```python
def get_signal_Hz(Hz, sample_rate, length_ts_sec):
    # Create a 1 second long list of length sample_rate containing the radian values of sampled points
    # on our sine wave
    ts1sec = list(np.linspace(0,np.pi*2*Hz,sample_rate))
    ## Multiply this list by length_ts_sec to get it to be the requested length
    ts = ts1sec*length_ts_sec
    # Return a list representing a sine wave of requested frequency and duration
    return(list(np.sin(ts)))
```

Now we can use get_signal_Hz to create our sound arrays representing the different analog numbers being dialed.

For 3 seconds of dialing '1': <br>
Two sine waves are created: one at 697Hz and one at 1209Hz. <br>
So we need to use <br>
get_signal_Hz(697, 4000, 3) and <br>
get_signal_Hz(1209, 4000, 3)

```python
sample_rate   = 4000
length_ts_sec = 3

ts1  = np.array(get_signal_Hz(697, sample_rate,length_ts_sec)) 
ts1 += np.array(get_signal_Hz(1209,sample_rate,length_ts_sec))
ts1  = list(ts1)
```

Now we have 1 second of silence. <br>
We just want an array of 4000 entries (our sampling rate) of [0]: <br>
[0] * 4000

```python
ts_silence = [0]*sample_rate*1
```

Lastly,

For 3 seconds of dialling '2': <br>
Two sine waves are created: one at 697Hz and one at 1336Hz. <br>
So we need to use <br>
get_signal_Hz(697, 4000, 3) and <br>
get_signal_Hz(1336, 4000, 3)

```python
sample_rate   = 4000
length_ts_sec = 3

ts2  = np.array(get_signal_Hz(697, sample_rate,length_ts_sec)) 
ts2 += np.array(get_signal_Hz(1336,sample_rate,length_ts_sec))
ts2  = list(ts2)
```

Now finally we add them together. This is why we converted the arrays into lists, so that we can combine them:

```python
ts = ts1 + ts_silence + ts2
```

We can use Audio from IPython.display to listen to time series audio files. 

```python
from IPython.display import Audio
Audio(ts, rate=sample_rate)
```

## Time Domain Plot
Now that we have generated our test data, let us plot it in the time domain. This will show us how the amplitude of the signal changes over time.

```python
# The total length is the total number of samples, len(ts), divided by the sampling rate, sample_rate
total_ts_sec = len(ts)/sample_rate
print("The total time series length = {} sec (N points = {}) ".format(total_ts_sec, len(ts)))

# Plotting figure
plt.figure(figsize=(20,3))
# This plots all the 28000 values in ts, the y value is the value of an element in ts,
# the x value corresponds to the index of the element
plt.plot(ts)

# For the y axis, the amplitude ticks are already correct by default

# For the x axis, we need to convert the element index order (the sample number) into seconds.
# Without this, it's 0-28000, with it, its 0 to 7
plt.xticks(np.arange(0,len(ts),sample_rate),
           np.arange(0,len(ts)/sample_rate,1))
# plt.xticks takes two parameters, the ticks in the original data, 
# and the ticks in the converted data. 
# The original data is between 0-28000.
# The converted data is between 0 and 28000/4000, 0 and 7 seconds. 

# Labelling axes and title:
plt.ylabel("Amplitude")
plt.xlabel("Time (second)")
plt.title("The total length of time series = {} sec, sample_rate = {}".format(len(ts)/sample_rate, sample_rate))
plt.show()
```

## Frequency Domain Plot

We will use a Discrete Fourier Transform for this task. <br>

Our fourier coefficient calculator function, get_xns, only calculates the Fourier coefficients up to the Nyquest limit. <br>
This video at https://www.youtube.com/watch?t=501&v=mkGsMWi_j4Q&feature=youtu.be provides more information about the Nyquest limit. <br>

The absolute value of each Fourier coefficient is also doubled to account for the symmetry of the Fourier coefficients around the Nyquest limit. <br>

The Nyquest Limit = sampling rate / 2. It is impossible to calculate frequencies of signals above this value. This tells us that we if want to fully calculate the frequencies of waves that form an audio file, we need a sampling rate double the highest frequency wave in said file.

First we need to code DFT:

```python
# This function takes an entry from a time series, Xs, along with the length of that time series,n,
# and calculates the fourier coefficients for it.
def get_xn(Xs,n):
    '''
    calculate the Fourier coefficient X_n of 
    Discrete Fourier Transform (DFT)
    '''
    L  = len(Xs)
    ks = np.arange(0,L,1)
    xn = np.sum(Xs*np.exp((1j*2*np.pi*ks*n)/L))/L
    return(xn)

# This function takes a time series and returns all the fourier coefficients for its entries
# upto the Nyquest limit.
def get_xns(ts):
    '''
    Compute Fourier coefficients only up to the Nyquest Limit Xn, n=1,...,L/2
    and multiply the absolute value of the Fourier coefficients by 2, 
    to account for the symetry of the Fourier coefficients above the Nyquest Limit. 
    '''
    mag = []
    L = len(ts)
    for n in range(int(L/2)): # Nyquest Limit
        mag.append(np.abs(get_xn(ts,n))*2)
    return(mag)
mag = get_xns(ts)
```

Now that we have the Fourier coefficients for our time series array, we can plot them to represent our audio using them:

```python
# Plotting figure
plt.figure(figsize=(20,3))
plt.plot(mag)

# For the y axis, the Fourier Cofficient ticks are already correct by default
# For the x axis, frequency in Hertz is NOT plotted. It's Fourier coefficient's frequency, k.

plt.xlabel("Frequency (k)")
plt.title("Two-sided frequency plot")
plt.ylabel("|Fourier Coefficient|")
plt.show()
```

The y axis is fine, but the x axis is not frequency in Hertz. We need to fix this in order to tell which waves are present in the audio.

We can convert a (by default) Fourier coefficient's frequency, k, into frequency in Hertz via

In our case, Sample Rate is 4000 and Total N of Sample Points is 28000.

```python
def get_Hz_scale_vec(ks,sample_rate,Npoints):
    freq_Hz = ks*sample_rate/Npoints
    freq_Hz  = [int(i) for i in freq_Hz ] 
    return(freq_Hz)


# Nxlim is the number of intervals to label along the xaxis
Nxlim = 30

# Creating ks and ksHz for xticks.
ks   = np.linspace(0,len(mag),Nxlim)
ksHz = get_Hz_scale_vec(ks,sample_rate,len(ts))

# Plotting
plt.figure(figsize=(20,3))
plt.plot(mag)
# Converting the x axis with xticks
plt.xticks(ks,ksHz)

# Labelling
plt.title("Frequency Domain")
plt.xlabel("Frequency (Hz)")
plt.ylabel("|Fourier Coefficient|")
plt.show()
```

From this, we can deduce a few things about our audio file. <br>
Firstly, there's a peak around 689Hz that's double in Fourier Coefficient value than the other signals. This must be our 697Hz sine waves, and is double the others because there are two of them. <br>
Secondly, we see our 1209Hz and 1336Hz peaks, one peak for each sine wave.

The frequency domain plot is great because it shows us the signal's underlying frequencies in Hz which we can interpret as we did. However, it is limited because it does not tell us anything about time (temporal information). E.g., when did the 1209Hz sine wave occur during the 7 seconds?

A frequency domain plot is created from the frequency dependent Fourier Coefficients. In order to display temporal information however, we'd need a plot created from the time dependent Fourier Coefficients. 

A spectrogram is one such way plot.

# Creating a spectrogram
Instead of doing DFT on the entire dataset, we can do it on subsets of it at a time and go across the data. This is doing DFT for a subwindow of the original time series data and sliding it across the signal. By doing so, we obtain the time-dependent Fourier coefficients.

Such a process is called short term DFT (SDFT). The name likely intuitive from it being DFT, but not on the whole time series at once, on small subsets (subwindows) instead.

A spectrogram is a way to plot time dependent Fourier Coefficients.

## Spectrogram Code
First there is a cell of code to create a spectrogram, then there is a cell of the code to plot it.

```python
def create_spectrogram(ts,NFFT,noverlap = None):
    '''
          ts: original time series
        NFFT: The number of data points used in each block for the DFT.
          Fs: the number of points sampled per second, so called sample_rate
    noverlap: The number of points of overlap between blocks. The default value is 128. 
    '''
    # By default, we set no overlap between the blocks.
    if noverlap is None:
        # For no overlap, we use half the number of data points per DFT block
        noverlap = NFFT/2
    noverlap = int(noverlap)
    
    # Randomly pick numbers between 0 and the length of the time series
    # NFFT-noverlap times
    starts  = np.arange(0,len(ts),NFFT-noverlap,dtype=int)
    
    # starts is an array, so we can do elementwise comparison
    # to remove windows that less than NFFT sample size
    starts  = starts[starts + NFFT < len(ts)]
    
    # A list of all the Fourier Coeffients in the time series for all windows
    xns = []
    for start in starts: # for each window
        # for each window, we apply a short term discrete fourier transform
        # to do so, we get the Fourier Coeffients of said window
        ts_window = get_xns(ts[start:start + NFFT]) 
        # append the window's Fourier Coeffients
        xns.append(ts_window)
    # specX is a transpose of xns
    specX = np.array(xns).T
    # as specX is an array, we can rescale it's (absolute) values easily
    spec = 10*np.log10(specX)
    
    # assert checks a condition
    # if it is True, then nothing happens, but if it's false, AssertionError occurs
    # spec.shape[1] should equal the length of starts, so we just check it
    assert spec.shape[1] == len(starts) 
    
    # return starts, containing window info,
    # and spec, the Fourier Coefficients
    return(starts,spec)

L = 256
noverlap = 84
starts, spec = create_spectrogram(ts,L,noverlap = noverlap)
```

```python
def plot_spectrogram(spec,ks,sample_rate, L, starts, mappable = None):
    # Create plot
    plt.figure(figsize=(20,8))
    # Plot the Fourier Coefficents
    plt_spec = plt.imshow(spec,origin='lower')

    # Creating the ticks for y
    # The number of intervals to plot on the axis
    Nyticks = 10
    # ks and ksHz for yticks
    ks      = np.linspace(0,spec.shape[0],Nyticks)
    ksHz    = get_Hz_scale_vec(ks,sample_rate,len(ts))
    # plotting
    plt.yticks(ks,ksHz)
    plt.ylabel("Frequency 10^3 (Hz)")

    ## Creating the ticks for x, involving manipulating the (non-Hertz frequency dependent) Fourier Coefficients.
    # The number of intervals to plot on the axis
    Nxticks = 10
    # ts_spec and ts_spec_sec for xticks
    ts_spec = np.linspace(0,spec.shape[1],Nxticks)
    ts_spec_sec  = ["{:4.2f}".format(i) for i in np.linspace(0,total_ts_sec*starts[-1]/len(ts),Nxticks)]
    # plotting
    plt.xticks(ts_spec,ts_spec_sec)
    plt.xlabel("Time (sec)")
    
    
    # Plotting labels/colours
    plt.title("Spectrogram L={} Spectrogram.shape={}".format(L,spec.shape))
    plt.colorbar(mappable,use_gridspec=True)
    plt.show()
    return(plt_spec)
plot_spectrogram(spec,ks,sample_rate,L,starts)
```

Looking at this, we can clearly see that the first 3 seconds had our 693Hz and 1209Hz sine waves, then we had our second of silence, and lastly our 3 seconds of 693Hz and 1336Hz sine waves!

## Spectrograms and Time/Frequency Uncertainty

Lastly we can modify the spectrogram to see how the relationship between time and frequency resolution change. They are an uncertainty principle. If we have high frequency resolution, our time resolution is low and vise versa. 

This manifests in spectrograms as blurry horizontal lines for frequency uncertainty. For time uncertainty, it manifests as blurry vertical lines, but it's hard to see here because our signal is so simple.

The following three plots show that as frequency resolution gets better (decreases), time resolution gets worse (increases). We can (should) see the horizontal line get narrower while the vertical line gets more blurry.

```python
plt_spec1 = None 
for iL, (L, bandnm) in enumerate(zip([150, 200, 400],["wideband","middleband","narrowband"])):
    print("{:20} time resolusion={:4.2f}sec, frequency resoulsion={:4.2f}Hz".format(bandnm,L/sample_rate,sample_rate/L))
    starts, spec = create_spectrogram(ts,L,noverlap = 1 )
    plt_spec = plot_spectrogram(spec,ks,sample_rate, L, starts, 
                                 mappable = plt_spec1)
    if iL == 0:
        plt_spec1 = plt_spec
```
