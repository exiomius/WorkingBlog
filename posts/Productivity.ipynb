# What is overworking? 
Overworking is expending too much time and energy on work that your brain and body slow down. This has effects on your productivity and happiness, and is not optimal for either. 

# Non-Linear relationship
In the past, humans often did 'non-brainy' tasks for work. Imagine working at a factory. Hours spent working is roughly proportional to units created. If working for one hour produces 10 items, then working for five hours produces 50. However, in modern life, especially as a student, most of not all of our work is not non-brainy, in fact, it's the complete opposite. This linear relationship between hours spent working at a desk and productivity loses it's feasibility.

# Why is overworking detrimental to productivity?

If you are overworked, many of your cognitive faculties lose their sharpness. Sure physically you get tired, which is more obvious, but cognitive faculties like emotional regulation, creative thought, self-esteem etc most often take a plunge as your mental health declines as you continue to overwork.

Creative thought decline, and its severe consequences on productivity, is often overlooked as a result of overworking. Intuitively, the tasks we are required to do will be done better if new ideas flow and are refined. Coding is a good example, a great programmer will implement fresh, creative, and inspiring ideas in their work. But there's also a less intuitive consequence. Creativity doesn't just improve things directly relating to the subject matter, but outside and around it. You need creativity, flexible thought, in order to effectively prioritise what work to do, and how to function efficiently.
